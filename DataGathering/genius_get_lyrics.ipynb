```python
import lyricsgenius as genius

import pandas as pd
import numpy as np
import re
import string
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('../audio_features.csv')[['name','artists']]
df['lyrics'] = [np.nan] * len(df)
df
```

```python
client_id = ''
client_secret = ''
client_access_token = ''
```

```python
genius = genius.Genius(client_access_token,
              response_format='plain',         # Format of response is plain text
              timeout=5, sleep_time=0.5, 
              remove_section_headers=True,     # Remove headers such as [Intro], [Verse], [Chorus]n etc.
              skip_non_songs=True,             # Skip items that are not songs
              verbose=True)
```

```python
#
for i in range(df.loc[~df['lyrics'].isnull()].index.values[-1]+1, len(df)):
    name_clean = re.split(' \(| - ', df['name'][i])[0]
    artist_clean = re.split(', | &', df['artists'][i])[0]
    song = genius.search_song(name_clean, artist=artist_clean)
    if not song:
        pass
    else:
        df['lyrics'][i] = song.lyrics.replace('\n', ' ').replace('EmbedShare URLCopyEmbedCopy','').replace('URLCopyEmbedCopy'.lower(),'').rstrip(string.digits)
```
