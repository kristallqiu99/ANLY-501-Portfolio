```python
import pandas as pd
import numpy as np
import requests
```

# Merge song release data and Billboard data

```python
song = pd.read_csv('song_release_ID_spotify.csv')[['name','artists','id','album.release_date']]
hit = pd.read_csv('billboard_ID_spotify.csv')[['name','artists','id','album.release_date','hit']]
```

```python
song
```

```python
hit
```

```python
merge1 = song.merge(hit, on=['id'], how='outer')
merge1
```

```python
merge2 = song.merge(hit, on=['name','artists'], how='outer')
merge2
```

*merge1* has more rows probably because the same song has different IDs in song release data and billboard data. Therefore, it is more reliable to use Spotify name and artists to merge. As a result, *merge2* has less rows.

```python
merge2.loc[merge2['id_x'].isnull(), ['id_x','album.release_date_x']] = \
merge2.loc[merge2['id_x'].isnull(), ['id_y','album.release_date_y']].values

merge2
```

```python
merge2 = merge2.drop(columns=['id_y','album.release_date_y'])
merge2.rename(columns={'id_x':'id', 'album.release_date_x':'release_date'}, inplace=True)
merge2.loc[merge2['hit'].isnull(),'hit'] = 0
```

```python
merge2.to_csv('total_song_merged.csv', index=False)
```

# Get audio features

```python
# initialize audio_feature dataframe to record audio features of tracks
audio_features = pd.DataFrame(columns = 
                              ['danceability','energy','key','loudness','mode','speechiness',
                               'acousticness','instrumentalness','liveness','valence','tempo',
                               'type','id','uri','track_href','analysis_url','duration_ms','time_signature'])
```

```python
CLIENT_ID = '6ced8cf12e0c470f9d9dd25a75d4ec26'
CLIENT_SECRET = '4ce366cebd7c42f2ac59bf96ca9f32b9'

AUTH_URL = 'https://accounts.spotify.com/api/token'

# POST
auth_response = requests.post(AUTH_URL, {
    'grant_type': 'client_credentials',
    'client_id': CLIENT_ID,
    'client_secret': CLIENT_SECRET,
})

# convert the response to JSON
auth_response_data = auth_response.json()

# save the access token
access_token = auth_response_data['access_token']
headers = {
    'Authorization': 'Bearer {token}'.format(token=access_token)
}

BASE_URL = 'https://api.spotify.com/v1/'

# max ids is 100
for i in range(0, len(merge2), 100):
    if not i+100 > len(merge2):
        ids = ','.join(merge2['id'][i:i+100])
    else:
        ids = ','.join(merge2['id'][i:])
        
    r = requests.get(BASE_URL + 'audio-features?ids=' + ids, headers=headers)
    
    audio_features = pd.concat([audio_features, pd.json_normalize(r.json()['audio_features'])],
                               ignore_index=True)
```

```python
full = pd.merge(merge2,
    audio_features[['danceability','energy','key','loudness','mode','speechiness',
                               'acousticness','instrumentalness','liveness','valence','tempo',
                               'id','duration_ms','time_signature']],
    how='left', on=['id'])
full
```
